```python
from pathlib import Path
import os
from scipy.io.wavfile import read as read_wav
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
import numpy as np
import parselmouth
import librosa as lib
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import seaborn as sn
```

PTAKI

```python
current_dir = Path.cwd() 
path_to_birds = current_dir / 'Nagrania/Birds/'
print(path_to_birds)

paths_to_bird_files = []
for directory, subdirs, files in os.walk(path_to_birds):
    paths_to_bird_files.extend([os.path.join(file) for file in files]) 

paths_to_bird_files.remove('.DS_Store')
print(paths_to_bird_files)

birds_signals = []
birds_samplerates = []

print(path_to_birds)

for i in range(0,len(paths_to_bird_files)):
    actual_path = path_to_birds/paths_to_bird_files[i]
    s, fs = lib.load(actual_path)
    #fs, s = read_wav(actual_path)
    if len(s.shape) > 1:
        s = s.mean(axis=1)
    birds_signals.append(s)
    birds_samplerates.append(fs)
```

```python
current_dir = Path.cwd() 
path_to_dogs = current_dir / 'Nagrania/Dogs/'
print(path_to_dogs)

paths_to_dog_files = []
for directory, subdirs, files in os.walk(path_to_dogs):
    paths_to_dog_files.extend([os.path.join(file) for file in files]) 

paths_to_dog_files.remove('.DS_Store')

dog_signals = []
dog_samplerates = []

for i in range(0,len(paths_to_dog_files)):
    actual_path = path_to_dogs/paths_to_dog_files[i]
    s, fs = lib.load(actual_path)

    if len(s.shape) > 1:
        s = s.mean(axis=1)
    dog_signals.append(s)
    dog_samplerates.append(fs)
```

```python
current_dir = Path.cwd() 
path_to_frogs = current_dir / 'Nagrania/frogs/'
print(path_to_frogs)

paths_to_frog_files = []
for directory, subdirs, files in os.walk(path_to_frogs):
    paths_to_frog_files.extend([os.path.join(file) for file in files]) 

paths_to_frog_files.remove('.DS_Store')

frog_signals = []
frog_samplerates = []

for i in range(0,len(paths_to_frog_files)):
    actual_path = path_to_frogs/paths_to_frog_files[i]
    s, fs = lib.load(actual_path)
  
    if len(s.shape) > 1:
        s = s.mean(axis=1)
    frog_signals.append(s)
    frog_samplerates.append(fs)   
```

INSEKTY

```python
current_dir = Path.cwd() 
path_to_insects = current_dir / 'Nagrania/Insect/'
print(path_to_insects)

paths_to_insect_files = []
for directory, subdirs, files in os.walk(path_to_insects):
    paths_to_insect_files.extend([os.path.join(file) for file in files]) 

paths_to_insect_files.remove('.DS_Store')

insect_signals = []
insect_samplerates = []

for i in range(0,len(paths_to_insect_files)):
    actual_path = path_to_insects/paths_to_insect_files[i]
    s, fs = lib.load(actual_path)

    if len(s.shape) > 1:
        s = s.mean(axis=1)
    insect_signals.append(s)
    insect_samplerates.append(fs)   
```

```python

'''3 pierwsze ptaki'''
for i in range(1, 4):
    #Spectrogramy
    f, t, bird_s = spectrogram(birds_signals[i-1], birds_samplerates[i-1])
    plt.figure()
    plt.ylim([0, 7000])
    plt.xlabel("Time, s")
    plt.ylabel("Frequency, Hz")
    plt.title(f'Bird nr {i} Spectrogram')
    plt.pcolormesh(t, f, bird_s, shading='gouraud', cmap='plasma')
```

```python
'''Trzy pierwsze insekty'''
for i in range(1, 4):
    #Spectrogramy
    f, t, bird_s = spectrogram(insect_signals[i-1], insect_samplerates[i-1])
    plt.figure()
    plt.ylim([0, 7000])
    plt.xlabel("Time, s")
    plt.ylabel("Frequency, Hz")
    plt.title(f'Insect nr {i} Spectrogram')
    plt.pcolormesh(t, f, bird_s, shading='gouraud', cmap='plasma')
```

```python
plt.plot(birds_signals[0])
```

```python
window = np.hamming(len(birds_signals[1]))
s = birds_signals[1] *window

fft_output = np.fft.fft(s)
freqs = np.fft.fftfreq(len(s), 1/birds_samplerates[1])
plt.figure(figsize=(10, 6))
plt.plot(freqs[:len(freqs)//2], 20* np.log10(np.abs(fft_output[:len(fft_output)//2])))
plt.xlabel('Częstotliwość (Hz)')
plt.ylabel('Amplituda')
plt.title('FFT sygnału dźwiękowego')
plt.grid(True)
plt.show()
```

```python
all_signals = birds_signals+dog_signals+frog_signals+insect_signals
all_samplerates = birds_samplerates+dog_samplerates+frog_samplerates+insect_samplerates
mfccs = []
sig_len = []
for signal in all_signals:
    mfcc_13_coeffs = lib.feature.mfcc(y=signal, sr=fs, n_mfcc=13)
    mfccs.append(mfcc_13_coeffs)
    sig_len.append(np.shape(mfcc_13_coeffs)[1])
prep_mfccs = []
for item in mfccs:
    item = item[:, :np.min(sig_len)]
    prep_mfccs.append(item)
mfccs = np.asarray(prep_mfccs)
```

```python
fig, ax = plt.subplots()
img = lib.display.specshow(mfccs[0], x_axis='time', ax=ax)
fig.colorbar(img, ax=ax) 
ax.set(title='MFCC')
```

```python
label_bird = ["Bird" for i in range(0, len(birds_signals))]
label_dog = ["Dog" for i in range(0, len(dog_signals))]
label_frog = ["Frog" for i in range(0, len(frog_signals))]
label_insect = ["Insect" for i in range(0, len(insect_signals))]


labels = label_bird+label_dog+label_frog+label_insect
```

```python
print(label_insect)
```

```python
print(labels)
```

```python

X_train, X_test, y_train, y_test = train_test_split(mfccs, labels, test_size=0.45, random_state=42, stratify=labels)

X_train = X_train.reshape(np.shape(X_train)[0],-1)
X_train.shape
X_test = X_test.reshape(np.shape(X_test)[0],-1)
```

```python
clf = KNeighborsClassifier(n_neighbors=4) #liczba sąsiadów nie musi być taka sama jak w zadaniu 1 - każde cechy 
#wymagają osobnego doboru liczby sąsiadów tak, by wyniki były jak najlepsze
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
preds
```

```python
matrix = confusion_matrix(y_test, preds)
```

```python
f1=f1_score(y_pred=preds,y_true=y_test,average='macro')
acc=accuracy_score(y_pred=preds,y_true=y_test)
print(f"Otrzymane metryki:\n"
      f"f1={f1}\n"
      f"accuracy={acc}")
```

```python
animals = ["Birds", "Dogs", "Frogs", "Insect"]
df_cm = pd.DataFrame(matrix, index = sorted(animals),
                  columns = sorted(animals))
plt.figure(figsize = (12,7))
plt.title('Confusion Matrix for KNN Classifier')
sn.heatmap(df_cm, annot=True)
```
